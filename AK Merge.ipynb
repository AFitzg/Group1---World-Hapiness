```python
score_GDP_2015 = happiness_metadata[["Score_2015","GDP_2015"]]
score_GDP_2015 = score_GDP_2015.rename(columns={"Score_2015":"Score","GDP_2015":"GDP"})
score_GDP_2016 = happiness_metadata[["Score_2016","GDP_2016"]]
score_GDP_2016 = score_GDP_2016.rename(columns={"Score_2016":"Score","GDP_2016":"GDP"})
score_GDP_2017 = happiness_metadata[["Score_2017","GDP_2017"]]
score_GDP_2017 = score_GDP_2017.rename(columns={"Score_2017":"Score","GDP_2017":"GDP"})
score_GDP_2018 = happiness_metadata[["Score_2018","GDP_2018"]]
score_GDP_2018 = score_GDP_2018.rename(columns={"Score_2018":"Score","GDP_2018":"GDP"})
score_GDP_2019 = happiness_metadata[["Score_2019","GDP_2019"]]
score_GDP_2019 = score_GDP_2019.rename(columns={"Score_2019":"Score","GDP_2019":"GDP"})
score_GDP = pd.concat([score_GDP_2015,score_GDP_2016,score_GDP_2017,score_GDP_2018,score_GDP_2019])
score_GDP
#Scatterplot and linear regression between happiness score and GDP for all the years
x_scatter = score_GDP["Score"]
y_scatter = score_GDP["GDP"]
score = score_GDP["Score"]
gdp = score_GDP["GDP"]
(slope_gdp, intercept_gdp, rvalue_gdp, pvalue_gdp, stderr_gdp) = linregress(score, gdp)
regress_values_gdp = score * slope_gdp + intercept_gdp
line_eq_gdp = "y = " + str(round(slope_gdp,2)) + "x + " + str(round(intercept_gdp,2))
plt2.scatter(x_scatter,y_scatter)
plt2.xlabel("Happiness Score")
plt2.ylabel("GDP per Capita")
plt2.title("Scatter Plot of Happiness Score vs GDP")
plt2.plot(score,regress_values_gdp,"r-")
plt2.annotate(line_eq_gdp,(5,1.0),fontsize=15,color="red")
plt2.show()
print("The R Squared value is : "+ str(rvalue_gdp ** 2))

r_gdp = st.pearsonr(score,gdp)
print("Correlation Coefficient for Happiness Score and GDP per Capita for countries is "+ str(r_gdp))
```
